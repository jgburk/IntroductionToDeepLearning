```r
set.seed(88888888)

library(scales)

ETA <<- 0.01

# input layer
NUM_FEATURES <<- 2
INPUT_WIDTH <<- NUM_FEATURES + 1 # (bias)

# hidden layer
HIDDEN_LAYER_WIDTH <<- 5
ANN_WIDTH <<- HIDDEN_LAYER_WIDTH + 1 # (bias)

# output layer
NUM_CLASSES <<- 3
OUTPUT_WIDTH <<- NUM_CLASSES

NUM_EPOCHS <- 3000

# ----------------               -                -----------
# num features + 1 (input axons) 6 (hidden_axons) num classes
# ----------------               -                -----------

# input layer
input_activation <<- matrix(data = 1,
                            nrow = INPUT_WIDTH,
                            ncol = 1)
input_axon_w <<- matrix(data = runif(n = HIDDEN_LAYER_WIDTH * INPUT_WIDTH),
                        nrow = INPUT_WIDTH,
                        ncol = HIDDEN_LAYER_WIDTH)
# hidden layer
hidden_deltas <<- matrix(data = 1,
                         nrow = HIDDEN_LAYER_WIDTH,
                         ncol = 1)
hidden_activation <<- matrix(data = 1,
                             nrow = ANN_WIDTH,
                             ncol = 1)
hidden_axon_w <<- matrix(data = runif(n = ANN_WIDTH * OUTPUT_WIDTH),
                         nrow = ANN_WIDTH,
                         ncol = OUTPUT_WIDTH)
# output layer
output_deltas <<- matrix(data = 1,
                         nrow = OUTPUT_WIDTH,
                         ncol = 1)
output_activation <<- matrix(data = 1,
                             nrow = OUTPUT_WIDTH,
                             ncol = 1)

activation_function <- function(x){
  1/(1 + exp(-(x)))
}

activation_function_derivative <- function(x){
  activation_function(x)*(1 - activation_function(x))
}

forward_prop <- function(training_sample_features){
  # input layer
  input_activation <<- c(training_sample_features,1)

  hidden_activation <<-
    c(activation_function(t(input_axon_w) %*% input_activation),1)

  output_activation <<-
    activation_function(t(hidden_axon_w) %*% hidden_activation)
}

encode_class <- function(sample_class){
  ann_class_encoding <- numeric(NUM_CLASSES)
  ann_class_encoding[sample_class] <- 1
  return(ann_class_encoding)
}

calc_err <- function(sample_class){
  ann_class_encoding <- encode_class(sample_class)
  output_soft_max <- output_activation / sum(output_activation)
  return(output_soft_max - ann_class_encoding)
}

backward_prop <- function(sample_class){
  output_deltas <<- calc_err(sample_class)

  hidden_deltas <<-
    hidden_axon_w %*%
    output_deltas *
    activation_function_derivative(hidden_activation)

  input_axon_w <<-
    input_axon_w -
    ETA *
    input_activation %*%
    t(hidden_deltas[1:HIDDEN_LAYER_WIDTH])

  hidden_axon_w <<-
    hidden_axon_w -
    ETA *
    hidden_activation %*%
    t(output_deltas)

  return(mean(abs(output_activation - encode_class(sample_class))))
}

train <- function(training_samples_features, training_samples_classes){
  error_mean <- c()
  for(x in 1:NUM_EPOCHS){
    error_sum <- 0
    for(i in 1:nrow(training_samples_features)){
      forward_prop(training_samples_features[i,])
      error_sum <- error_sum + backward_prop(training_samples_classes[i])
    }
    error_mean <- c(error_mean,error_sum/nrow(training_samples_features))
  }
  plot(error_mean,
       xlab="Epoch",
       ylab="Error",
       main="Mean Training Error each Epoch")
}

predict0 <- function(test_sample_features){
  forward_prop(test_sample_features)
  prediction <- output_activation
  #print(output_activation)
  return(which(prediction == max(prediction)))
}

predict_n <- function(test_samples_features){
  ps <- c()
  for(i in 1:nrow(test_samples_features)){
    ps <- c(ps,predict0(test_samples_features[i,]))
  }
  return(ps)
}

samples_f <- matrix(
  c(0,0,
    0,.2,
    .3,.3,
    .3,.3,
    .3,.1,
    .6,.15,
    .6,.2,
    .6,.3,
    .6,.1,
    .6,.35,
    .8,.8,
    .8,.25,
    .8,.1,
    .8,.3,
    .8,.2,
    .8,.4),
  nrow = 16,
  ncol = 2,
  byrow = TRUE)

samples_c <- matrix(
  c(2,2,2,2,1,1,1,1,1,1,3,3,3,3,3,3),
  nrow = 16,
  ncol = 1,
  byrow = TRUE
)

norm_samples_f <- scales::rescale(samples_f, to = c(0,1))
train(norm_samples_f, samples_c)

plot(norm_samples_f,col=samples_c,pch=samples_c * 2,
     xlab="Column 1 Data",
     ylab="Column 2 Data",
     main="True Labels")

pred_train <- predict_n(norm_samples_f)
n_wrong <- sum(pred_train != samples_c)
n_total <- length(samples_c)
accuracy <- (1 - (n_wrong / n_total)) * 100
plot(norm_samples_f,col=pred_train,pch=pred_train * 2,
     xlab="Column 1 Data",
     ylab="Column 2 Data",
     main=paste("Predicted Labels: ",
                n_total - n_wrong," of ",n_total,
                " (",accuracy,"%) correct",sep=""))
if(n_wrong == 1){
  points(x=norm_samples_f[which(pred_train != samples_c),][1],
       y=norm_samples_f[which(pred_train != samples_c),][2],
       pch=1,
       cex=5)
}

rand_points <<- matrix(data = runif(n = 2 * 100),
                       nrow = 100,
                       ncol = 2)
norm_rand_points <- scales::rescale(rand_points, to = c(0,1))
pred_classes <- predict_n(norm_rand_points)
plot(norm_rand_points,col=pred_classes,pch=pred_classes * 2,
     xlab="Column 1 Dimension",
     ylab="Column 2 Dimension",
     main="Decision Boundary")
```

```r
set.seed(88888888)

library(scales)

DATA_DIR <- "./" #"/home/jgburk/PycharmProjects/IntroductionToDeepLearning/"
OUTPUT_DATA_FN <- "transposed.Rds"

NUM_TRAIN_IT <<- 500
ETA <<- 0.01 #step 2
RAND_POINTS <<- 500 # displays decision boundary
MCE_THRESH <<- 0.05 # maximum classwise error
RNA_SEQ_ANALYSIS <<- 1

# input layer
NUM_FEATURES <<- FEATURE_COUNT
INPUT_WIDTH <<- NUM_FEATURES + 1 # (bias)

# hidden layers
HIDDEN_LAYER_DEPTH <<- 3 #step 1
HIDDEN_CORE_DEPTH <<- HIDDEN_LAYER_DEPTH - 1 # special dims for final hidden layer
HIDDEN_LAYER_WIDTH <<- 8 # step 0
ANN_WIDTH <<- HIDDEN_LAYER_WIDTH + 1 # (bias)

# output layer
NUM_CLASSES <<- 7
OUTPUT_WIDTH <<- NUM_CLASSES

MAX_INIT <- .5
MIN_INIT <- -.5

# input layer
input_activation <<- matrix(data = 1,
                            nrow = INPUT_WIDTH, #includes bias
                            ncol = 1)
input_axon_w <<- matrix(data = runif(n = HIDDEN_LAYER_WIDTH * INPUT_WIDTH, min = MIN_INIT, max = MAX_INIT),
                        nrow = INPUT_WIDTH,
                        ncol = HIDDEN_LAYER_WIDTH)
# hidden core layers
hidden_core_deltas <<- matrix(data = 1,
                              nrow = ANN_WIDTH,
                              ncol = HIDDEN_CORE_DEPTH)
hidden_core_activation <<- matrix(data = 1,
                                  nrow = HIDDEN_LAYER_WIDTH,
                                  ncol = HIDDEN_CORE_DEPTH)
hidden_core_axon_w <<- array(data = runif(n =  HIDDEN_LAYER_WIDTH * ANN_WIDTH * HIDDEN_CORE_DEPTH, min = MIN_INIT, max = MAX_INIT),
                             dim = c(HIDDEN_LAYER_WIDTH,
                                     ANN_WIDTH,
                                     HIDDEN_CORE_DEPTH))

# hidden final layer
hidden_final_deltas <<- matrix(data = 1,
                               nrow = ANN_WIDTH,
                               ncol = 1)
hidden_final_activation <<- matrix(data = 1,
                                   nrow = HIDDEN_LAYER_WIDTH,
                                   ncol = 1)
hidden_final_axon_w <<- matrix(data = runif(n = ANN_WIDTH * OUTPUT_WIDTH, min = MIN_INIT, max = MAX_INIT),
                               nrow = ANN_WIDTH,
                               ncol = OUTPUT_WIDTH)
# output layer
output_deltas <<- matrix(data = 1,
                         nrow = OUTPUT_WIDTH,
                         ncol = 1)
output_activation <<- matrix(data = 1,
                             nrow = OUTPUT_WIDTH,
                             ncol = 1)

activation_function <- function(x){
  1/(1 + exp(-(x)))
}

activation_function_derivative <- function(x){
  activation_function(x)*(1 - activation_function(x))
}

forward_prop <- function(training_sample_features){
  # input layer
  input_activation <<- c(training_sample_features,1) # input_activation 17

  if(HIDDEN_CORE_DEPTH > 0){
    hidden_core_activation[,1] <<- # hidden_core_activation 30 x 9
      activation_function(t(input_axon_w) %*% input_activation) # input_axon_w 17 x 30

    if(HIDDEN_CORE_DEPTH > 1){
      for(core_layer in 2:HIDDEN_CORE_DEPTH){
        hidden_core_activation[,core_layer] <<- # hidden_core_activation 30 x 9
          activation_function(hidden_core_axon_w[,,core_layer - 1] %*% # hidden_core_axon_w 30 x 31 x 9
                                c(hidden_core_activation[,core_layer - 1],1)) # hidden_core_activation 30 x 9
      }
    }

    hidden_final_activation <<- # hidden_final_activation 30 x 1
      activation_function(hidden_core_axon_w[,,HIDDEN_CORE_DEPTH] %*% # hidden_core_axon_w 30 x 31 x 9
                            c(hidden_core_activation[,HIDDEN_CORE_DEPTH],1)) # hidden_core_activation 30 x 9
  }else{
    hidden_final_activation <<- # hidden_final_activation 30 x 1
      activation_function(t(input_axon_w) %*% # input_axon_w 17 x 30
                            input_activation) # input_activation 17
  }

  output_activation <<- # output_activation 7 x 1
    activation_function(t(hidden_final_axon_w) %*% # hidden_final_axon_w 31 x 7
                          c(hidden_final_activation,1)) # hidden_final_activation 30 x 1
}

encode_class <- function(sample_class){ # one-hot encoding
  ann_class_encoding <- numeric(NUM_CLASSES)
  ann_class_encoding[sample_class] <- 1
  return(ann_class_encoding)
}

calc_err <- function(sample_class){
  ann_class_encoding <- encode_class(sample_class)
  output_soft_max <- output_activation / sum(output_activation)
  return(output_soft_max - ann_class_encoding)
}

backward_prop <- function(sample_class){
  output_deltas <<- calc_err(sample_class) #output_deltas 7 x 1

  hidden_final_deltas <<-        # hidden_final_deltas 31 x 1
    hidden_final_axon_w %*%      # hidden_final_axon_w 31 x 7 (delTotal/delW)
    output_deltas *              # output_deltas 7 x 1 (delETotal/delOut)
    activation_function_derivative(c(hidden_final_activation[1:HIDDEN_LAYER_WIDTH],1)) # hidden_final_activation 30 x 1 (delOut/delInput)

  if(HIDDEN_CORE_DEPTH > 0){
    hidden_core_deltas[,HIDDEN_CORE_DEPTH] <<-        # hidden_core_deltas 31 x 9
      (t(hidden_core_axon_w[,,HIDDEN_CORE_DEPTH]) %*%     # hidden_core_axon_w 30 x 31 x 9
         hidden_final_deltas[1:HIDDEN_LAYER_WIDTH]) *                           # hidden_final_deltas 31 x 1
      activation_function_derivative(c(hidden_core_activation[1:HIDDEN_LAYER_WIDTH,HIDDEN_CORE_DEPTH],1)) # hidden_core_activation 30 x 9

    if(HIDDEN_CORE_DEPTH > 1){
      for(core_layer in (HIDDEN_CORE_DEPTH - 1):1){
        hidden_core_deltas[,core_layer] <<-           # hidden_core_deltas 31 x 9
          (t(hidden_core_axon_w[,,core_layer]) %*%        # hidden_core_axon_w 30 x 31 x 9
             hidden_core_deltas[1:HIDDEN_LAYER_WIDTH,core_layer + 1]) *       # hidden_core_deltas 31 x 9
          activation_function_derivative(c(hidden_core_activation[1:HIDDEN_LAYER_WIDTH,core_layer],1)) # hidden_core_activation 30 x 9
      }
    }

    input_axon_w_jacobian <- input_activation %*% # input_activation 17
      t(hidden_core_deltas[1:HIDDEN_LAYER_WIDTH,1]) # hidden_core_deltas 31 x 9

    input_axon_w <<- # input_axon_w 17 x 30
      input_axon_w - # input_axon_w 17 x 30
      ETA *
      input_axon_w_jacobian

    if(HIDDEN_CORE_DEPTH > 1){
      for(core_layer in 1:(HIDDEN_CORE_DEPTH - 1)){
        hidden_core_axon_w_jacobian <- hidden_core_activation[,core_layer] %*% # hidden_core_activation 30 x 9
          t(hidden_core_deltas[,core_layer + 1]) # hidden_core_deltas 31 x 9

        hidden_core_axon_w[,,core_layer] <<- # hidden_core_axon_w 30 x 31 x 9
          hidden_core_axon_w[,,core_layer] - # hidden_core_axon_w 30 x 31 x 9
          ETA *
          hidden_core_axon_w_jacobian
      }
    }

    hidden_core_axon_w_jacobian <- hidden_core_activation[,HIDDEN_CORE_DEPTH] %*% # hidden_core_activation 30 x 9
      t(hidden_final_deltas) # hidden_final_deltas 31 x 1

    hidden_core_axon_w[,,HIDDEN_CORE_DEPTH] <<-     # hidden_core_axon_w 30 x 31 x 9
      hidden_core_axon_w[,,HIDDEN_CORE_DEPTH] -     # hidden_core_axon_w 30 x 31 x 9
      ETA *
      hidden_core_axon_w_jacobian

  }else{
    input_axon_w_jacobian <- input_activation %*% # input_activation 17
      t(hidden_final_deltas[1:HIDDEN_LAYER_WIDTH]) # hidden_final_deltas 31 x 1

    input_axon_w <<- # input_axon_w 17 x 30
      input_axon_w - # input_axon_w 17 x 30
      ETA *
      input_axon_w_jacobian
  }

  hidden_final_axon_w_jacobian <- c(hidden_final_activation,1) %*% # hidden_final_activation 30 x 1
    t(output_deltas) # output_deltas 7 x 1

  hidden_final_axon_w <<- # hidden_final_axon_w 31 x 7
    hidden_final_axon_w - # hidden_final_axon_w 31 x 7
    ETA *
    hidden_final_axon_w_jacobian

  return(mean(abs(output_activation - encode_class(sample_class))))
}

GLOB_ERR_1 <<- c()
GLOB_ERR_2 <<- c()
GLOB_ERR_3 <<- c()
GLOB_ERR_4 <<- c()
GLOB_ERR_5 <<- c()
GLOB_ERR_6 <<- c()
GLOB_ERR_7 <<- c()
GLOB_MAX_CLASSWISE_ERR <<- c()

train <- function(training_samples_features, training_samples_classes){
  error_sum <- c()
  classwise_errors <- c()
  mean_error_sum <- c()
  for(x in 1:NUM_TRAIN_IT){
    error1 <- c()
    error2 <- c()
    error3 <- c()
    error4 <- c()
    error5 <- c()
    error6 <- c()
    error7 <- c()
    for(i in 1:nrow(training_samples_features)){
      #i <- ceiling(runif(n=1,min=0,max=nrow(training_samples_features)))
      forward_prop(training_samples_features[i,])
      error <- backward_prop(training_samples_classes[i])
      error_sum <- c(error_sum,error)
      classwise_errors <- c(classwise_errors,error)

      if(training_samples_classes[i] == 1){
        error1 <- c(error1,error)
      }else if(training_samples_classes[i] == 2){
        error2 <- c(error2,error)
      }else if(training_samples_classes[i] == 3){
        error3 <- c(error3,error)
      }else if(training_samples_classes[i] == 4){
        error4 <- c(error4,error)
      }else if(training_samples_classes[i] == 5){
        error5 <- c(error5,error)
      }else if(training_samples_classes[i] == 6){
        error6 <- c(error6,error)
      }else if(training_samples_classes[i] == 7){
        error7 <- c(error7,error)
      }
    }
    GLOB_ERR_1 <<- c(GLOB_ERR_1,mean(error1))
    GLOB_ERR_2 <<- c(GLOB_ERR_2,mean(error2))
    GLOB_ERR_3 <<- c(GLOB_ERR_3,mean(error3))
    GLOB_ERR_4 <<- c(GLOB_ERR_4,mean(error4))
    GLOB_ERR_5 <<- c(GLOB_ERR_5,mean(error5))
    GLOB_ERR_6 <<- c(GLOB_ERR_6,mean(error6))
    GLOB_ERR_7 <<- c(GLOB_ERR_7,mean(error7))

    mce <- max(classwise_errors) # maximum classwise error
    GLOB_MAX_CLASSWISE_ERR <<- c(GLOB_MAX_CLASSWISE_ERR,mce)
    print(paste("MCE ",x,": ",mce,sep=""))
    #zeta <- (mce / (2 * 3^(1/2)))
    mean_error_sum <- c(mean_error_sum,rep(mean(classwise_errors),nrow(training_samples_features)))
    #plot(error_sum, col = "black",main="Individual (Black) & Mean (Red) Errors")#,col=seq(1:nrow(training_samples_features)),pch = seq(1:nrow(training_samples_features)))
    #points(mean_error_sum, col="red")
    classwise_errors <- c() #clear classwise errors
    if (mce < MCE_THRESH){
      print(paste("mce < ",MCE_THRESH," achieved (mce = ",mce,"). Stopping...",sep=""))
      break
    }
  }
  plot(error_sum, col = "black",main="Individual (Black) & Mean (Red) Errors")#,col=seq(1:nrow(training_samples_features)),pch = seq(1:nrow(training_samples_features)))
  points(mean_error_sum, col="red")
}

predict1 <- function(test_sample_features){
  forward_prop(test_sample_features)
  prediction <- output_activation
  return(which(prediction == max(prediction)))
}

predict_n <- function(test_samples_features){
  ps <- c()
  for(i in 1:nrow(test_samples_features)){
    z <- predict1(test_samples_features[i,])
    if(length(z) > 2){
      print(z)
      z <- z[1]
    }
    ps <- c(ps,z)
  }
  print(length(ps))
  return(ps)
}

sample_data <- readRDS(paste(DATA_DIR,OUTPUT_DATA_FN,sep=""))
samples_f <- as.matrix(sample_data[,1:NUM_FEATURES])
samples_c <- as.matrix(sample_data[,NUM_FEATURES+1])
norm_samples_f <- matrix(nrow = nrow(samples_f),ncol = ncol(samples_f))
for(i in 1:ncol(samples_f)){
  norm_samples_f[,i] <- scales::rescale(samples_f[,i], to=c(0,1))
}

train(norm_samples_f, samples_c)
plot(norm_samples_f[,1:2],col=samples_c,pch=samples_c,main="True Labels")
pred_train <- predict_n(norm_samples_f)
plot(norm_samples_f[,1:2],col=pred_train,pch=pred_train,main="Predicted Labels")
rand_points <<- matrix(data = runif(n = NUM_FEATURES * RAND_POINTS),
                       nrow = RAND_POINTS,
                       ncol = NUM_FEATURES)
norm_rand_points <- scales::rescale(rand_points, to = c(0,1))
pred_classes <- predict_n(norm_rand_points)
plot(norm_rand_points[,1:2],col=pred_classes,pch=pred_classes,main="Decision Boundary")
training_misclass_rate <- sum(samples_c != pred_train) / length(pred_train)
print(paste("Train Misclass Rate: ",training_misclass_rate,sep=""))
plot(GLOB_ERR_1)
plot(GLOB_ERR_2)
plot(GLOB_ERR_3)
plot(GLOB_ERR_4)
plot(GLOB_ERR_5)
plot(GLOB_ERR_6)
plot(GLOB_ERR_7)
plot(GLOB_MAX_CLASSWISE_ERR)

hist(samples_c)

library(pROC)

mroc <- pROC::multiclass.roc(response=as.numeric(samples_c),predictor=as.numeric(pred_train))
for(roc in 1:length(mroc$rocs)){
  pROC::plot.roc(mroc$rocs[[roc]],main=paste("AUC = ",
                                             round(auc(mroc$rocs[[roc]]),digits=3),
                                             " for Targets ",
                                             mroc$rocs[[roc]]$levels[1],
                                             " and ",
                                             mroc$rocs[[roc]]$levels[2],
                                             sep=""))
}
```

```r
set.seed(88888888)

library(magrittr)
library(scales)
library(nnet)

DATA_DIR <- "./" #"/home/jgburk/PycharmProjects/IntroductionToDeepLearning/"
OUTPUT_DATA_FN <- "transposed.Rds"
NUM_FEATURES <<- FEATURE_COUNT
RAND_POINTS <<- 500 # displays decision boundary

sample_data <- readRDS(paste(DATA_DIR,OUTPUT_DATA_FN,sep=""))

samples_f <- as.matrix(sample_data[,1:NUM_FEATURES])
samples_c <- as.matrix(sample_data[,NUM_FEATURES+1])
norm_samples_f <- matrix(nrow = nrow(samples_f),ncol = ncol(samples_f))
for(i in 1:ncol(samples_f)){
  norm_samples_f[,i] <- scales::rescale(samples_f[,i], to=c(0,1))
}

nn_df <- data.frame(norm_samples_f)
nn_df$y <- as.factor(samples_c)

nn_model <- nnet::nnet(formula = y~.,
                       data=nn_df,
                       size=8,
                       maxitr=500)

nn_train_pred <- predict(object=nn_model,newdata=nn_df[,1:NUM_FEATURES])
pred_vec <- c()
for(i in 1:nrow(nn_train_pred)){
  pred_vec <- c(pred_vec,which(nn_train_pred[i,] == max(nn_train_pred[i,])))
}
plot(nn_df[,1:2],col=pred_vec,pch=pred_vec,main="nnet Predicted Labels")
nn_training_misclass_rate <- sum(as.numeric(nn_df$y) != pred_vec) / length(pred_vec)
print(paste("nnet Train Misclass Rate: ",nn_training_misclass_rate,sep=""))
rand_points <<- matrix(data = runif(n = NUM_FEATURES * RAND_POINTS),
                       nrow = RAND_POINTS,
                       ncol = NUM_FEATURES)
norm_rand_points <- scales::rescale(rand_points, to = c(0,1))
norm_rand_points_df <- data.frame(norm_rand_points)
colnames(norm_rand_points_df) <- colnames(nn_df[,1:NUM_FEATURES])
nn_pred_classes <- predict(object=nn_model,newdata=norm_rand_points_df)
pred_vec2 <- c()
for(i in 1:nrow(nn_pred_classes)){
  pred_vec2 <- c(pred_vec2,which(nn_pred_classes[i,] == max(nn_pred_classes[i,])))
}
plot(norm_rand_points[,1:2],col=pred_vec2,pch=pred_vec2,main="nnet Decision Boundary")

library(pROC)
mroc <- pROC::multiclass.roc(response=as.numeric(nn_df$y),predictor=as.numeric(pred_vec))
for(roc in 1:length(mroc$rocs)){
 pROC::plot.roc(mroc$rocs[[roc]],main=paste("nnet AUC = ",
                                            round(auc(mroc$rocs[[roc]]),digits=3),
                                            " for Targets ",
                                            mroc$rocs[[roc]]$levels[1],
                                            " and ",
                                            mroc$rocs[[roc]]$levels[2],
                                            sep=""))
}
```
